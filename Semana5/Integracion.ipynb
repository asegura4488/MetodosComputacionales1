```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
f = lambda x: np.sin(x)
x = np.linspace(0, np.pi, 20)
y = f(x)
```

```python
plt.plot(x, y,'.-',color='k')
plt.show()
```

```python
N = 6
x1 = np.linspace(0, np.pi, N+1)
y1 = f(x1)
yp = np.interp(x1,x1,y1)
print(yp)
#print(y1)
```

```python
plt.plot(x,y,color='k')
plt.plot(x1,yp,'o',color='r')
plt.fill_between(x,y,color='r',alpha=0.5)
plt.fill_between(x1,y1,color='b',alpha=0.5)
plt.show()
```

```python
class Integrator:

  def __init__(self, x,f):

    self.x = x
    self.y = f(self.x)
    self.h = self.x[1]-self.x[0]

    self.Integral = 0.
```

```python

#Clase hija
class Trapezoid(Integrator):

  def __init__(self,x,f):
    Integrator.__init__(self,x,f)

  def GetIntegral(self):

    self.Integral = 0.

    self.Integral += 0.5*self.h*(self.y[0]+self.y[-1])

    self.Integral += self.h*np.sum( self.y[1:-1] )

    return self.Integral
```

```python
class Simpson(Integrator):

  def __init__(self,x,f):
    Integrator.__init__(self,x,f)

  def GetIntegral(self):

    self.Integral = (self.h/3)*( self.y[0] + 4*np.sum( self.y[1:-1:2]) + 2*np.sum( self.y[2:-2:2]) + self.y[-1])

    return self.Integral
```

```python
Integrador1 = Trapezoid(x1,f)
Integrador1
```

```python
Integrador1.GetIntegral()
```

```python
integrate.trapezoid(f(x1),x1)
```
