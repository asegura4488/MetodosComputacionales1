```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)
x = np.linspace(0, np.pi, 20)
y = f(x)
```

```python
plt.plot(x, y,'.-',color='k')
plt.show()
```

```python
N = 10
x1 = np.linspace(0, np.pi, N+1)
y1 = f(x1)
yp = np.interp(x1,x1,y1)
```

```python
plt.plot(x,y,color='k')
plt.plot(x1,yp,'o',color='r')
plt.fill_between(x,y,color='r',alpha=0.5)
plt.fill_between(x1,y1,color='b',alpha=0.5)
plt.show()
```
