```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Files = ['2025/Metodos20251.xlsx']
df = pd.read_excel(Files[0])
```

```python
df.keys()
```

```python
# Por el momento
df['Complementaria'] = 4.0
```

```python
dfparciales = df[['Parcial 1','Parcial 2','Parcial 3','Parcial 4','Bonos']]
#dfparciales
```

```python
df = df.assign( Parciales = dfparciales.sum(axis=1)/4.)
```

```python
# ponderación acordada
df = df.assign( NotaFinal = np.round(0.75*(df.Parciales) + 0.25*df.Complementaria,2) ) 
df
```

```python
dffinal = df[['Last Name','First Name', 'NotaFinal']]
dffinal
#dffinal.sort_values(by='NotaFinal')
```

```python
dffinal.to_excel('Metodos1.xlsx',index=False,float_format="%.2f")
```

```python
percentiles = np.percentile(dffinal.NotaFinal,[0,25,50,75,100])



fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

ax.set_xticklabels(['Métodos computacionales 1'], rotation=1, fontsize=15)

ax.boxplot(dffinal.NotaFinal)
ax.set_ylabel('Grades', fontsize=15)

for i in range(len(percentiles)):
    ax.text(1+0.1,percentiles[i],np.round(percentiles[i],2),color='r')
```
