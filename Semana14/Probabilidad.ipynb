```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GetExperiment(N = 20):

  N = int(N)
  freq = np.zeros(N)
  NTimes = 0

  for i in range(int(N)):

    d1 = np.random.randint(1,7)
    if d1 == 2:
      NTimes +=1

  return NTimes/N
```

```python
GetExperiment(N=500)
```

```python
n = np.logspace(1,5,7)
freq = np.zeros_like(n)
for i in range(len(n)):
  freq[i] = GetExperiment(n[i])
```

```python
plt.plot(freq)
plt.axhline(y=1/6,color='r')
```

```python
def GetExperiment(N = 20):

  N = int(N)
  freq = np.zeros(N)

  for i in range(int(N)):

    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)

    freq[i] = d1+d2

  return freq
```

```python
freq = GetExperiment(N=10000)
#freq
```

```python
plt.hist(freq)
```

```python
x = np.linspace(2,13,12)
x
h, bins = np.histogram(freq,bins=x)
w = np.diff(bins)
w
```

```python
plt.bar(bins[:-1],h,color='b',width=w,ec='k')
```
