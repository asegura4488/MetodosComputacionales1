```python
# Derivacion
## Existe la función subyacente
## La función subyacente es "suave"
```

```python
def function(x):
  return np.sin(x)
```

```python
x = np.linspace(0, 2*np.pi, 20)
y = function(x)
# Derivation Step
h = x[1] - x[0]
print(h)
```

```python
def RightDerivative(f,x,h):

  d = (f(x+h) - f(x))/h

  return d
```

```python
def LeftDerivative(f,x,h):

  d = (f(x) - f(x-h))/h

  return d
```

```python
def CentralDerivative(f,x,h):

  d = (f(x+h) - f(x-h))/(2*h)

  return d
```

```python
Right = RightDerivative(function,x,h)
Left = LeftDerivative(function,x,h)
Central = CentralDerivative(function,x,h)
```

```python
#plt.plot(x,y)
fig = plt.figure(figsize=(4,4))
#plt.scatter(x,y,color='k')
plt.scatter(x,np.cos(x),color='k')
plt.scatter(x,Right,color='b')
plt.scatter(x,Left,color='r')
plt.scatter(x,Central,color='g')
```

```python
from copy import Error
# Error Local
ErrorRight = np.abs( np.cos(x) - Right )
ErrorLeft = np.abs( np.cos(x) - Left )
ErrorCentral = np.abs( np.cos(x) - Central )

plt.scatter(x,ErrorRight, color='b')
plt.scatter(x,ErrorLeft, color='r')
plt.scatter(x,ErrorCentral, color='g')
```

```python
def SecondDerivative(f,x,h):

  d = (f(x+h) - 2*f(x) + f(x-h))/h**2

  return d
```

```python
Second = SecondDerivative(function,x,h)
```
